```python
def quant(u1,z1):
    ac = np.quantile(u1['count'], np.arange(0,1001)/1000) #assembly
    gc = np.quantile(z1['count'], np.arange(0,1001)/1000) #gene
    pc = pc = pd.DataFrame(zip(ac,gc))
    pc[3] = (pc[0] - pc[0].min())/(pc[0].max() - pc[0].min())
    pc[4] = (pc[1] - pc[1].min())/(pc[1].max() - pc[1].min())
    return pc[[3,4]]
```

```python
#2d quantile plot with 10000 resolution. increased granularity
def quant(u1,z1):
    ac = np.quantile(u1['count'], np.arange(0,10001)/10000) #assembly
    gc = np.quantile(z1['count'], np.arange(0,10001)/10000) #gene
    pc = pc = pd.DataFrame(zip(ac,gc))
    pc[3] = (pc[0] - pc[0].min())/(pc[0].max() - pc[0].min())
    pc[4] = (pc[1] - pc[1].min())/(pc[1].max() - pc[1].min())
    return pc[[3,4]]

pc1 = quant(u1,z1).iloc[9500:]
pc9 = quant(u9,z9).iloc[9500:]
pc5 = quant(u5,z5).iloc[9500:]
pc10 = quant(u10,z10).iloc[9500:]

fig, ax = plt.subplots(nrows=2, ncols=2, figsize=(14, 14))

sns.kdeplot(data=pc1, x=4, y=3, fill=True, ax = ax[0,0])
ax[0,0].title.set_text("Arthropoda: 1013 BUSCOs tested in 2422 Assemblies")
ax[0,0].set(xlabel = "Upper 5% Gene Quantile", ylabel = "Upper 5% Assembly quantile")

sns.kdeplot(data=pc9, x=4, y=3, fill=True, ax = ax[1,0])
ax[1,0].title.set_text("Vertebrata: 3354 BUSCOs tested in 2685 Assemblies")
ax[1,0].set(xlabel = "Upper 5% Gene Quantile", ylabel = "Upper 5% Assembly quantile")

sns.kdeplot(data=pc5, x=4, y=3, fill=True, ax = ax[0,1])
ax[0,1].title.set_text("Eudicots: 2326 BUSCOs tested in 980 Assemblies")
ax[0,1].set(xlabel = "Upper 5% Gene Quantile", ylabel = "Upper 5% Assembly quantile")

sns.kdeplot(data=pc10, x=4, y=3, fill=True, ax = ax[1,1])
ax[1,1].title.set_text("Viridiplantae: 425 BUSCOs tested in 1327 Assemblies")
ax[1,1].set(xlabel = "Upper 5% Gene Quantile", ylabel = "Upper 5% Assembly quantile")
```
